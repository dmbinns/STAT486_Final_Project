```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


## Load Data
df = pd.read_csv("../data/utah_fitness_combined.csv")

# -----------------------------
# BASIC INFO
# -----------------------------
print("Shape of dataset:", df.shape)

# -----------------------------
# NUMERIC SUMMARY STATS
# -----------------------------
numeric_cols = [
    'review_count', 'rating', 'median_income',
    'distance', 'total_pop', 'median_age', 'pct_bachelors'
]

# Summary statistics
summary_stats = df[numeric_cols].describe()
print(summary_stats)

# Mean + SD specifically
mean_sd = df[numeric_cols].agg(['mean', 'std'])
print("\nMean and Standard Deviation:\n", mean_sd)

# -----------------------------
# CATEGORICAL SUMMARY
# -----------------------------
print("\nCategory Counts:")
category_counts = df['category'].value_counts()
print(category_counts)

# -----------------------------
# TARGET VARIABLE DISTRIBUTION
# -----------------------------
plt.figure()
df['review_count'].hist(bins=30)
plt.title("Distribution of Review Count")
plt.xlabel("Review Count")
plt.ylabel("Frequency")
plt.show()

# -----------------------------
# RATING DISTRIBUTION
# -----------------------------
plt.figure()
df['rating'].hist(bins=20)
plt.title("Distribution of Ratings")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()

# -----------------------------
# INCOME VS REVIEW COUNT
# -----------------------------
plt.figure()
plt.scatter(df['median_income'], df['review_count'])
plt.xlabel("Median Income")
plt.ylabel("Review Count")
plt.title("Income vs Review Count")
plt.show()

# -----------------------------
# CATEGORY VS REVIEW COUNT (BOXPLOT)
# -----------------------------
plt.figure()
df.boxplot(column='review_count', by='category', rot=45)
plt.title("Review Count by Category")
plt.suptitle("")  # removes default title
plt.xlabel("Category")
plt.ylabel("Review Count")
plt.show()

# -----------------------------
# CORRELATION MATRIX
# -----------------------------
corr_matrix = df[numeric_cols].corr()
print("\nCorrelation Matrix:\n", corr_matrix)
```
